```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)

plt.title("Распределение возраста пользователей")
plt.xlabel('Возраст')
plt.ylabel('Частота')
```

```python
sns.jointplot(x='Age', y='Area Income', data=df, kind='scatter', height=8)
```

```python
sns.jointplot(x='Daily Time Spent on Site', y='Daily Internet Usage', data=df, kind='hex', height=8)
```

```python
sns.pairplot(df, hue='Clicked on Ad', palette='Set1')
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Hour'] = df['Timestamp'].dt.hour
df['DayOfWeek'] = df['Timestamp'].dt.dayofweek
df['Month'] = df['Timestamp'].dt.month
df.head()
```

```python
plt.figure(figsize=(12, 6))
df.groupby(['Hour', 'Clicked on Ad']).size().unstack().plot(kind='bar', stacked=False, figsize=(12, 6))

plt.title('Распределение кликов по часам дня')
plt.xlabel('Час дня')
plt.ylabel('Количество')
plt.legend(['Не кликнули', 'Кликнули'])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.groupby(['DayOfWeek', 'Clicked on Ad']).size().unstack().plot(kind='bar', figsize=(12, 6))

plt.title('Распределение кликов по дням недели')
plt.xlabel('День недели (0=Понедельник 6=Воскресенье)')
plt.legend(['Не кликнули', 'Кликнули'])
plt.xticks(rotation=0)
plt.show()
```
