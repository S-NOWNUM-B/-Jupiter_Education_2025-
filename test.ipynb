```python
import matplotlib.pyplot as plt

# Данные о длительности событий (в часах) за спринт (2 недели)
# Данные взяты из раздела 1.2 и 3.3
events = {
    'Sprint Planning': 4.0,       # 4 часа (начало спринта)
    'Daily Scrum': 2.5,           # 15 мин * 10 дней = 150 мин = 2.5 часа
    'Backlog Refinement': 2.0,    # 2 часа (середина спринта)
    'Sprint Review': 2.0,         # 2 часа (конец спринта)
    'Sprint Retrospective': 1.5   # 1.5 часа (конец спринта)
}

labels = list(events.keys())
sizes = list(events.values())
colors = ['#4a90e2', '#66bb6a', '#ffa726', '#ab47bc', '#ef5350']
explode = (0.05, 0, 0, 0, 0)  # Немного выдвигаем самый большой сектор (Planning)

# Создание графика
fig, ax = plt.subplots(figsize=(9, 6))

wedges, texts, autotexts = ax.pie(sizes, explode=explode, labels=labels, colors=colors,
                                  autopct='%1.1f%%', shadow=False, startangle=140,
                                  pctdistance=0.85, textprops={'fontsize': 10})

# Делаем "пончик" (Donut chart) для современного вида
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Настройка текста внутри секторов
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_weight('bold')

# Заголовок и подпись общего времени
total_hours = sum(sizes)
ax.set_title('Распределение времени на Scrum Events', fontsize=14, fontweight='bold')
ax.text(0, 0, f'Всего:\n{total_hours} часов\n(на коммуникацию)', ha='center', va='center', fontsize=12, fontweight='bold', color='#555')

plt.tight_layout()
plt.show()
```
