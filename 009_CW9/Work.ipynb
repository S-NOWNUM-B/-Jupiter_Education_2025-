Создание отдельных файлов по регионам

```python
os.makedirs('reports/blacklist_ent', exist_ok=True)

blacklist = df[df['qedu_passed_ent'] < 0.5]
blacklist.to_excel('reports/blacklist_ent/TOTAL_blacklist.xlsx', index=False)

print(f"Всего школ в чёрном списке: {len(blacklist)}")
print(f"Это {len(blacklist)/len(df)*100:.2f}% от общего числа\n")

for region in df['region'].unique():
    region_bl = blacklist[blacklist['region'] == region]
    if len(region_bl) > 0:
        filename = f"reports/blacklist_ent/blacklist_{region.replace(' ', '_')}.xlsx"
        region_bl.to_excel(filename, index=False)
        print(f"✓ {region}: {len(region_bl)} школ")

print("\nЧёрные списки сохранены")
```

Чёрный список школ с низкими результатами ЕНТ

```python
os.makedirs('reports/material_problems', exist_ok=True)

problems = df[
    (df['drinking_water_presence'].isin(['нет', 'no'])) |
    (df['load_ratio'] > 1.5) |
    (df['deficit_places_2018'] > 0)
]

problems.to_excel('reports/material_problems/TOTAL_problems.xlsx', index=False)
print(f"Школ с проблемами: {len(problems)}\n")

for region in df['region'].unique():
    region_prob = problems[problems['region'] == region]
    if len(region_prob) > 0:
        filename = f"reports/material_problems/problems_{region.replace(' ', '_')}.xlsx"
        region_prob.to_excel(filename, index=False)
        print(f"✓ {region}: {len(region_prob)} школ")
```

Список школ с проблемами материальной обеспеченности

```python
os.makedirs('reports/top_teachers', exist_ok=True)

if 'eng_informatics_teachers' in df.columns:
    df['total_eng'] = (df['eng_informatics_teachers'].fillna(0) + 
                       df['eng_phys_teachers'].fillna(0) +
                       df['eng_chem_teachers'].fillna(0) + 
                       df['eng_bio_teachers'].fillna(0))
    
    top50 = df.nlargest(50, 'total_eng')[['region', 'total_eng']]
    top50.to_excel('reports/top_teachers/TOP_50_english.xlsx', index=False)
    
    print("Топ-10 школ по учителям с английским:")
    print(top50.head(10).to_string(index=False))
    
    for spec, col in [('informatics', 'eng_informatics_teachers'),
                      ('physics', 'eng_phys_teachers'),
                      ('chemistry', 'eng_chem_teachers'),
                      ('biology', 'eng_bio_teachers')]:
        top20 = df.nlargest(20, col)[['region', col]]
        top20.to_excel(f'reports/top_teachers/TOP_20_{spec}.xlsx', index=False)
        print(f"✓ {spec}: топ-20 сохранён")
```

Топ школ по учителям со знанием английского языка

```python
city_cols = [col for col in df.columns if 'city' in col.lower() or 'location' in col.lower()]

if city_cols:
    city_col = city_cols[0]
    
    city_stats = df.groupby(city_col).agg({
        'region': 'first',
        'ID': 'count',
        'total_students_2018': 'sum',
        'qedu_passed_ent': 'mean',
        'load_ratio': 'mean',
        'gdp': 'first'
    }).reset_index()
    
    city_stats.columns = ['Населённый_пункт', 'Регион', 'Количество_школ', 
                          'Всего_учеников', 'Средний_ЕНТ', 'Загруженность', 'ВВП']
    
    city_stats = city_stats.sort_values('Количество_школ', ascending=False)
    city_stats.to_excel('reports/statistics_by_city.xlsx', index=False)
    
    print(f"Статистика по {len(city_stats)} населённым пунктам")
    print("\nТоп-10:")
    print(city_stats.head(10).to_string(index=False))
```

```python
# Проверяем наличие файла GDP by region.xlsx
import os

if os.path.exists('GDP by region.xlsx'):
    # Загружаем данные о ВВП
    gdp_df = pd.read_excel('GDP by region.xlsx')
    print("Файл GDP by region.xlsx успешно загружен!")
    print(f"\nРазмер датасета ВВП: {gdp_df.shape}")
    print("\nПервые строки:")
    display(gdp_df.head())
    
    # Объединяем с основным датасетом
    # Предполагаем, что в обоих датасетах есть колонка с названием региона
    print("\nКолонки в GDP датасете:", gdp_df.columns.tolist())
else:
    print("Файл GDP by region.xlsx не найден!")
    print("Создадим синтетические данные о ВВП для демонстрации...")
    
    # Создаем синтетические данные о ВВП по регионам
    regions = df['region'].unique() if 'region' in df.columns else []
    
    if len(regions) > 0:
        gdp_data = {
            'region': regions,
            'gdp': np.random.randint(1000000, 10000000, size=len(regions))
        }
        gdp_df = pd.DataFrame(gdp_data)
        print("\nСоздан синтетический датасет ВВП:")
        display(gdp_df.head())
```

## 5. Визуализация и анализ взаимосвязей ВВП и показателей школ

```python
# Анализ взаимосвязи ВВП и качества образования (баллы ЕНТ)
if 'qedu_passed_ent' in df.columns and 'gdp' in df.columns:
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # 1. Гистограмма распределения баллов ЕНТ по регионам
    df_grouped = df.groupby('region')[['gdp', 'qedu_passed_ent']].mean().reset_index()
    
    axes[0, 0].barh(df_grouped['region'], df_grouped['qedu_passed_ent'], color='steelblue')
    axes[0, 0].set_xlabel('Средний процент сдавших ЕНТ')
    axes[0, 0].set_title('Средний процент сдавших ЕНТ по регионам')
    axes[0, 0].grid(axis='x', alpha=0.3)
    
    # 2. Scatter plot: ВВП vs баллы ЕНТ
    axes[0, 1].scatter(df_grouped['gdp'], df_grouped['qedu_passed_ent'], s=100, alpha=0.6, color='coral')
    axes[0, 1].set_xlabel('ВВП региона')
    axes[0, 1].set_ylabel('Средний % сдавших ЕНТ')
    axes[0, 1].set_title('Взаимосвязь ВВП региона и результатов ЕНТ')
    axes[0, 1].grid(alpha=0.3)
    
    # Добавляем линию тренда
    z = np.polyfit(df_grouped['gdp'], df_grouped['qedu_passed_ent'], 1)
    p = np.poly1d(z)
    axes[0, 1].plot(df_grouped['gdp'], p(df_grouped['gdp']), "r--", alpha=0.8, label='Тренд')
    axes[0, 1].legend()
    
    # 3. Гистограмма ВВП по регионам
    axes[1, 0].barh(df_grouped['region'], df_grouped['gdp'], color='lightgreen')
    axes[1, 0].set_xlabel('ВВП региона')
    axes[1, 0].set_title('ВВП по регионам')
    axes[1, 0].grid(axis='x', alpha=0.3)
    
    # 4. Корреляция
    correlation = df_grouped['gdp'].corr(df_grouped['qedu_passed_ent'])
    axes[1, 1].text(0.5, 0.5, f'Корреляция ВВП и\nрезультатов ЕНТ:\n\n{correlation:.4f}', 
                    ha='center', va='center', fontsize=16, 
                    bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    axes[1, 1].axis('off')
    axes[1, 1].set_title('Коэффициент корреляции')
    
    plt.tight_layout()
    plt.show()
    
    print(f"Корреляция между ВВП региона и результатами ЕНТ: {correlation:.4f}")
```

Анализируем взаимосвязь между ВВП региона и качеством образования (результаты ЕНТ)

```python
# Анализ взаимосвязи ВВП и материальной базы школ
# Проверяем наличие колонок, связанных с материальной базой
material_columns = [col for col in df.columns if any(keyword in col.lower() 
                    for keyword in ['toilet', 'туалет', 'repair', 'ремонт', 'heating', 'отопление'])]

print("Колонки, связанные с материальной базой:")
for col in material_columns[:10]:  # Выводим первые 10
    print(f"  - {col}")

# Статистический анализ
if 'gdp' in df.columns:
    # Создаем сводную таблицу по регионам
    region_stats = df.groupby('region').agg({
        'gdp': 'first',
        'qedu_passed_ent': 'mean',
        'total_students_2018': 'sum',
        'load_ratio': 'mean' if 'load_ratio' in df.columns else 'count'
    }).reset_index()
    
    print("\n" + "="*60)
    print("Статистика по регионам:")
    print("="*60)
    print(region_stats.to_string(index=False))
    
    # Квартили
    print("\n" + "="*60)
    print("Квартили результатов ЕНТ:")
    print("="*60)
    print(df['qedu_passed_ent'].quantile([0.25, 0.5, 0.75]))
```

### 7.3. Список школ с проблемами материальной обеспеченности

```python
# Список школ с проблемами материальной обеспеченности
material_dir = os.path.join(output_dir, 'material_problems')
os.makedirs(material_dir, exist_ok=True)

print("Поиск школ с проблемами материальной обеспеченности")
print("=" * 60)

# Критерии для определения проблем с материальной базой:
# 1. Питьевая вода отсутствует
# 2. Высокая загруженность (load_ratio > 1.5)
# 3. Туалеты на улице (если есть такая колонка)

problems_schools = df.copy()

# Фильтруем по критериям
conditions = []

# 1. Проблемы с питьевой водой
if 'drinking_water_presence' in df.columns:
    no_water = problems_schools['drinking_water_presence'].isin(['нет', 'no', 'No'])
    conditions.append(no_water)
    print(f"Школ без питьевой воды: {no_water.sum()}")

# 2. Высокая загруженность
if 'load_ratio' in df.columns:
    high_load = problems_schools['load_ratio'] > 1.5
    conditions.append(high_load)
    print(f"Школ с высокой загруженностью (>1.5): {high_load.sum()}")

# 3. Дефицит мест
if 'deficit_places_2018' in df.columns:
    has_deficit = problems_schools['deficit_places_2018'] > 0
    conditions.append(has_deficit)
    print(f"Школ с дефицитом мест: {has_deficit.sum()}")

# Объединяем условия (школы с хотя бы одной проблемой)
if conditions:
    combined_condition = conditions[0]
    for condition in conditions[1:]:
        combined_condition = combined_condition | condition
    
    problem_schools = problems_schools[combined_condition]
    
    print(f"\nВсего школ с проблемами: {len(problem_schools)}")
    print(f"Это составляет {len(problem_schools)/len(df)*100:.2f}% от общего числа школ")
    
    # Сохраняем общий список
    problem_schools.to_excel(os.path.join(material_dir, 'TOTAL_material_problems.xlsx'), index=False)
    
    # Создаем отдельные списки для каждого региона
    print("\nПроблемные школы по регионам:")
    print("=" * 60)
    
    for region in regions:
        region_problems = problem_schools[problem_schools['region'] == region]
        
        if len(region_problems) > 0:
            safe_region_name = region.replace(' ', '_').replace('/', '_')
            filename = f"problems_{safe_region_name}.xlsx"
            filepath = os.path.join(material_dir, filename)
            
            region_problems.to_excel(filepath, index=False)
            
            # Статистика по типам проблем
            stats = []
            if 'drinking_water_presence' in df.columns:
                no_water_count = region_problems['drinking_water_presence'].isin(['нет', 'no', 'No']).sum()
                if no_water_count > 0:
                    stats.append(f"без воды: {no_water_count}")
            
            if 'load_ratio' in df.columns:
                high_load_count = (region_problems['load_ratio'] > 1.5).sum()
                if high_load_count > 0:
                    stats.append(f"перегружены: {high_load_count}")
            
            stats_str = ", ".join(stats) if stats else "разные проблемы"
            print(f"✓ {region}: {len(region_problems)} школ ({stats_str})")
    
    print("\n" + "=" * 60)
    print(f"Отчёты сохранены в папку: {material_dir}")
else:
    print("Не найдено подходящих колонок для анализа материальной базы")
```

Создаем списки школ с проблемами материальной обеспеченности (без воды, перегруженность, дефицит мест)

### 7.4. Топ школ по учителям, владеющим английским языком

```python
# Топ школ по учителям со знанием английского языка
top_teachers_dir = os.path.join(output_dir, 'top_teachers')
os.makedirs(top_teachers_dir, exist_ok=True)

print("Анализ учителей, владеющих английским языком и специализациями")
print("=" * 60)

# Поиск колонок с учителями по специальностям
eng_columns = [col for col in df.columns if 'eng' in col.lower() and 'teacher' in col.lower()]

print("\nНайденные колонки:")
for col in eng_columns:
    print(f"  - {col}")

# Проверяем наличие нужных колонок
if 'eng_informatics_teachers' in df.columns:
    # Создаем сумму всех учителей со знанием английского
    df['total_eng_teachers'] = 0
    
    for col in ['eng_informatics_teachers', 'eng_phys_teachers', 
                'eng_chem_teachers', 'eng_bio_teachers']:
        if col in df.columns:
            df['total_eng_teachers'] += df[col].fillna(0)
    
    # Топ школ по общему количеству учителей с английским
    top_eng = df.nlargest(50, 'total_eng_teachers')[['region', 'school_type', 'total_eng_teachers',
                                                       'eng_informatics_teachers', 'eng_phys_teachers',
                                                       'eng_chem_teachers', 'eng_bio_teachers']].copy()
    
    top_eng.to_excel(os.path.join(top_teachers_dir, 'TOP_50_english_teachers.xlsx'), index=False)
    
    print(f"\nТоп-10 школ по учителям со знанием английского:")
    print("=" * 60)
    for idx, row in top_eng.head(10).iterrows():
        print(f"{row['region']}: {int(row['total_eng_teachers'])} учителей")
    
    # Топ школ по каждой специальности
    specialties = {
        'informatics': 'eng_informatics_teachers',
        'physics': 'eng_phys_teachers',
        'chemistry': 'eng_chem_teachers',
        'biology': 'eng_bio_teachers'
    }
    
    print("\nТоп школ по специальностям:")
    print("=" * 60)
    
    for spec_name, col_name in specialties.items():
        if col_name in df.columns:
            top_spec = df.nlargest(20, col_name)[['region', 'school_type', col_name]].copy()
            
            filename = f'TOP_20_{spec_name}_english.xlsx'
            top_spec.to_excel(os.path.join(top_teachers_dir, filename), index=False)
            
            total = top_spec[col_name].sum()
            print(f"✓ {spec_name.capitalize()}: сохранено топ-20 школ (всего учителей: {int(total)})")
    
    print("\n" + "=" * 60)
    print(f"Отчёты сохранены в папку: {top_teachers_dir}")
else:
    print("Колонки с учителями по специальностям не найдены")
```

Создаем топ школ по количеству учителей со знанием английского языка (общий и по специальностям)

### 7.5. Статистика по населённым пунктам

```python
# Статистика по населённым пунктам
# Проверяем наличие колонки с населенным пунктом
location_columns = [col for col in df.columns if any(keyword in col.lower() 
                    for keyword in ['city', 'город', 'location', 'населен', 'locality', 'place'])]

print("Найденные колонки с населёнными пунктами:")
for col in location_columns[:10]:
    print(f"  - {col}")

# Используем первую подходящую колонку
if location_columns:
    city_col = location_columns[0]
    print(f"\nИспользуем колонку: {city_col}")
    
    # Проверяем наличие колонок для анализа
    internet_cols = [col for col in df.columns if 'internet' in col.lower() or 'интернет' in col.lower()]
    
    print(f"\nКолонки с интернетом: {internet_cols[:3]}")
    
    # Создаем агрегированную статистику по населённым пунктам
    city_stats = df.groupby(city_col).agg({
        'region': 'first',
        'ID': 'count',  # Количество школ
        'total_students_2018': 'sum',  # Общее количество учеников
        'qedu_passed_ent': 'mean',  # Средний процент сдавших ЕНТ
        'load_ratio': 'mean' if 'load_ratio' in df.columns else 'count',
    }).reset_index()
    
    # Переименовываем колонки
    city_stats.columns = ['Населённый_пункт', 'Регион', 'Количество_школ', 
                          'Всего_учеников', 'Средний_результат_ЕНТ', 'Средняя_загруженность']
    
    # Добавляем статистику по интернету
    if internet_cols:
        internet_col = internet_cols[0]
        
        # Процент школ с доступом в интернет
        internet_stats = df.groupby(city_col).apply(
            lambda x: (x[internet_col].isin(['да', 'yes', 'Yes', 'Да', 1, True]).sum() / len(x) * 100)
        ).reset_index()
        internet_stats.columns = ['Населённый_пункт', 'Процент_школ_с_интернетом']
        
        city_stats = city_stats.merge(internet_stats, on='Населённый_пункт', how='left')
    
    # Проверяем наличие хотя бы одной школы без доступа к питьевой воде
    if 'drinking_water_presence' in df.columns:
        water_stats = df.groupby(city_col).apply(
            lambda x: 'Да' if any(x['drinking_water_presence'].isin(['нет', 'no', 'No'])) else 'Нет'
        ).reset_index()
        water_stats.columns = ['Населённый_пункт', 'Есть_школы_без_воды']
        
        city_stats = city_stats.merge(water_stats, on='Населённый_пункт', how='left')
    
    # Добавляем дополнительную статистику
    # Средний ВВП региона
    if 'gdp' in df.columns:
        gdp_stats = df.groupby(city_col)['gdp'].first().reset_index()
        gdp_stats.columns = ['Населённый_пункт', 'ВВП_региона']
        city_stats = city_stats.merge(gdp_stats, on='Населённый_пункт', how='left')
    
    # Количество учителей
    teacher_cols = [col for col in df.columns if 'total' in col.lower() and 'teacher' in col.lower()]
    if teacher_cols:
        teacher_stats = df.groupby(city_col)[teacher_cols[0]].sum().reset_index()
        teacher_stats.columns = ['Населённый_пункт', 'Всего_учителей']
        city_stats = city_stats.merge(teacher_stats, on='Населённый_пункт', how='left')
    
    # Сортируем по количеству школ
    city_stats = city_stats.sort_values('Количество_школ', ascending=False)
    
    # Сохраняем отчёт
    city_stats.to_excel(os.path.join(output_dir, 'statistics_by_city.xlsx'), index=False)
    
    print(f"\n{'='*60}")
    print(f"Статистика создана для {len(city_stats)} населённых пунктов")
    print(f"{'='*60}")
    print(f"\nТоп-10 населённых пунктов по количеству школ:")
    print(city_stats.head(10).to_string(index=False))
    
    print(f"\n{'='*60}")
    print(f"Отчёт сохранён: {os.path.join(output_dir, 'statistics_by_city.xlsx')}")
    
    # Инсайты
    print(f"\n{'='*60}")
    print("ИНСАЙТЫ:")
    print(f"{'='*60}")
    
    if 'Процент_школ_с_интернетом' in city_stats.columns:
        worst_internet = city_stats.nsmallest(5, 'Процент_школ_с_интернетом')
        print(f"\n1. Населённые пункты с наихудшим доступом к интернету:")
        for _, row in worst_internet.iterrows():
            if row['Количество_школ'] > 1:  # Только если более 1 школы
                print(f"   - {row['Населённый_пункт']}: {row['Процент_школ_с_интернетом']:.1f}% школ с интернетом")
    
    if 'Средний_результат_ЕНТ' in city_stats.columns:
        best_ent = city_stats.nlargest(5, 'Средний_результат_ЕНТ')
        print(f"\n2. Населённые пункты с лучшими результатами ЕНТ:")
        for _, row in best_ent.head(5).iterrows():
            print(f"   - {row['Населённый_пункт']}: {row['Средний_результат_ЕНТ']:.2%} средний результат")
    
    if 'Есть_школы_без_воды' in city_stats.columns:
        no_water_count = (city_stats['Есть_школы_без_воды'] == 'Да').sum()
        print(f"\n3. В {no_water_count} населённых пунктах есть школы без доступа к питьевой воде")
    
    if 'Средняя_загруженность' in city_stats.columns:
        overloaded = city_stats[city_stats['Средняя_загруженность'] > 1.5]
        print(f"\n4. {len(overloaded)} населённых пунктов имеют среднюю загруженность школ > 1.5")
        if len(overloaded) > 0:
            print("   Наиболее перегруженные:")
            for _, row in overloaded.nlargest(3, 'Средняя_загруженность').iterrows():
                print(f"   - {row['Населённый_пункт']}: {row['Средняя_загруженность']:.2f}")
                
else:
    print("Не найдена колонка с населёнными пунктами")
```
